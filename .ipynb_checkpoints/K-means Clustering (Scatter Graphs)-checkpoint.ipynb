```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random as rd
import matplotlib.pyplot as plt
from math import sqrt

df = pd.read_csv("/home/steph/mental-health-research-proj/student_mental_health.csv")
df_original = pd.read_csv("/home/steph/mental-health-research-proj/student_mental_health_original.csv")
```

```python
df_original.drop(['Timestamp'], axis = 1, inplace = True)

df_original.info()
df_original.head()
```

```python
df.drop(['Timestamp'], axis = 1, inplace = True)

df.info()
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X = df

X['What_is_your_course'] = le.fit_transform(X['What_is_your_course'])

df.info()
df.head()

# y = le.transform(y)
```

```python
# WIP
plt.rcParams["figure.figsize"] = (40, 40)
plt.bar("What_is_your_course", "Do_you_have_Depression", data = df, color = "blue")
plt.xlabel("What_is_your_course")
plt.xticks(rotation = 90)
plt.ylabel("Do_you_have_Depression")
plt.title("Course & Depression")
plt.show()
```
