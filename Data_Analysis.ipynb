```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/steph/mental-health-research-proj/student_mental_health.csv')
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
plt.hist(df['Age'],color='cornflowerblue')
plt.title("Age distribution");
```

```python
df.rename(columns = {'Choose_your_gender': 'gender'}, inplace = True)
```

```python
plt.figure(figsize=(12,6))
plt.title("Gender Distribution")
g = plt.pie(df.gender.value_counts(), explode=(0.025,0.025), labels=df.gender.value_counts().index, colors=['lightpink','lightgreen'],autopct='%1.1f%%', startangle=180);
plt.legend()
plt.show()
# 0 is male and 1 is female
```

```python
plt.figure(figsize=(10,10))
sns.countplot(df['Your_current_year_of_Study'],hue=df['gender'])
plt.title("Current year of study");
```

```python
plt.figure(figsize=(10,10))
sns.countplot(df['Do_you_have_Anxiety'],hue=df['Do_you_have_Depression'])
plt.title("Students with anxiety and depression");
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(y="Do_you_have_Anxiety", hue="gender", data=df)
plt.title("Anxiety by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(y="Do_you_have_Depression", hue="gender", data=df)
plt.title("Depression by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Do_you_have_Anxiety", hue="Your_current_year_of_Study", data=df)
plt.title("Anxiety by study year")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Do_you_have_Depression", hue="Your_current_year_of_Study", data=df)
plt.title("Depression by study year")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Do_you_have_Panic_attack", hue="What_is_your_CGPA", data=df)
plt.title("Panic attack by CGPA")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Do_you_have_Depression", hue="What_is_your_CGPA", data=df)
plt.title("Depression by CGPA")
plt.show()
```
